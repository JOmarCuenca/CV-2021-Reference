# Referencia de la API de Matplotlib para uso posterior y que no tenga que ir a la documentacion y empezar desde cero

## Importaciones necesarias para poder correr todo el notebook

```python
import matplotlib.pyplot as pplot
%matplotlib inline
%config Completer.use_jedi = False
from random import shuffle,randint
import numpy as np
```

### Rendereado de una grafica simple 

```python
x = list(range(10))
y = list(range(50,60))
shuffle(y)
```

```python
pplot.xlabel("Day")
pplot.ylabel("Temperature")
pplot.title("Weather")
pplot.plot(x,y,color="orange",linewidth=3,linestyle="dotted")
```

```python
pplot.plot(x,y,"g*--") # color green con marcador *--
```

###  Rendereado de Multiples graficas en una sola grafica, con leyenda y un grid

```python
days = list(range(1,8))
max_t= list(range(25,32))
shuffle(max_t)
min_t= list(range(13,20))
shuffle(min_t)
avg_t = [(max_t[x]+min_t[x])/2 for x in range(len(max_t))]
```

```python
pplot.title("Weather")
pplot.ylabel("Temperature in C")
pplot.xlabel("Days of the Week")
pplot.plot(days,max_t,label="Max")
pplot.plot(days,min_t,label="Min")
pplot.plot(days,avg_t,label="Average")
pplot.legend(loc="best",shadow=True)
pplot.grid()
```

### Rendereado de una barra de graficas 

```python
company = ["Google","AWS","MicroSoft","FaceBook"]
revenue = [randint(400,1000) for x in range(len(company))]
profit = [randint(0,300) for x in range(len(company))]
xpos = np.arange(len(company))
```

```python
pplot.xticks(xpos,company)
pplot.bar(xpos-.2,revenue,label="Revenue",width=.4)
pplot.bar(xpos+.2,profit,label="Profit",width=.4)
pplot.legend()
```

### Rendereado de una grafica de barras horizontal

```python
pplot.yticks(xpos,company)
pplot.barh(xpos-.2,revenue,label="Revenue",height=.4)
pplot.barh(xpos+.2,profit,label="Profit",height=.4)
pplot.legend()
```

### Rendereado de un histograma (Barras pero organizadas)

```python
length = range(10)
blood_sugar_men = [randint(75,160) for x in length]
blood_sugar_woman = [randint(75,160) for x in length]
```

```python
pplot.xlabel("sugar range")
pplot.ylabel("# of patients")
pplot.hist([blood_sugar_men,blood_sugar_woman],bins=[80,100,125,150],rwidth=.5,label=["Men","Women"])
pplot.legend()
```

```python
pplot.xlabel("sugar range")
pplot.ylabel("# of patients")
pplot.hist([blood_sugar_men,blood_sugar_woman],bins=[80,100,125,150],rwidth=.5,label=["Men","Women"],
          orientation="horizontal")
pplot.legend()
```

### Una grafica de pie y como exportarlo

```python
exp_labels = ["Videogames","Beer","Movies","Girlfriend"]
exp_vals = [randint(100,1600) for x in range(len(exp_labels))]
avg = sum(exp_vals)/len(exp_vals)
```

```python
pplot.axis("equal") # just in case the pie isn't a perfect circle
pplot.pie(exp_vals,
          labels=exp_labels,
          radius=2,
          autopct="%0.2f%%",
          shadow=True,
          explode=[.2 if x >= avg else 0 for x in exp_vals],
          startangle=100
         )
# pplot.show() # dont show the values the pplot library prints by default
pplot.savefig("piechart.png",bbox_inches="tight",pad_inches=.5,transparent=False)
```
